```python
plt.figure(figsize = (10,5))
plt.bar(list(fifa_20['nationality'].value_counts()[0:5].keys()), list(fifa_20['nationality'].value_counts()[0:5]),
        color = ['skyblue','red','green','orange','pink'])
plt.show()
```

#### Getting salary of each player in euros

```python
player_sal = fifa_20[['short_name','wage_eur']]
```

```python
player_sal.head()
```

#### sort the players by their salary

```python
player_sal = player_sal.sort_values(by = ['wage_eur'],ascending = False)
```

```python
player_sal.head()
```

#### Visualizing salaries of each player

```python

plt.figure(figsize = (10,5))
plt.bar(list(player_sal['short_name'])[0:5],list(player_sal['wage_eur'])[0:5],color = ['skyblue','red','green','orange','pink'])
plt.show()
```
